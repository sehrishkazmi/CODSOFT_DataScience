```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# KNN - nearest neighbour
```

```python
iris_data = pd.read_csv("IRIS.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
iris_data['species_encoded'] = encoder.fit_transform(iris_data['species'])
```

```python
iris_data
```

```python
# Creating a dictionary to store the names of species
iris_species = {0: 'Iris-setosa', 1: 'Iris-versicolor', 2: 'Iris-virginica'}
iris_species[0]
```

```python
X = iris_data.drop(['species_encoded','species'], axis=1) # axis=1 for columns
y = iris_data['species_encoded']
```

```python
# Splitting Train and Test Sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=53)
```

```python
# Check the shapes
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Applying Standard Scaler when using KNN Algorithm
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test) 
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
